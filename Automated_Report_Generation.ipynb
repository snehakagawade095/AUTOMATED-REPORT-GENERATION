```python
import pandas as pd
from reportlab.platypus import SimpleDocTemplate, Paragraph, Table, TableStyle
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.lib import colors
data = pd.read_csv("50_Startups.csv")
average_profit = data["Profit"].mean()
max_profit = data["Profit"].max()
min_profit = data["Profit"].min()
pdf = SimpleDocTemplate("Startup_Profit_Report.pdf")
styles = getSampleStyleSheet()
elements = []
elements.append(Paragraph("Startup Profit Analysis Report", styles["Title"]))
elements.append(Paragraph(f"Average Profit: {average_profit:.2f}", styles["Normal"]))
elements.append(Paragraph(f"Maximum Profit: {max_profit:.2f}", styles["Normal"]))
elements.append(Paragraph(f"Minimum Profit: {min_profit:.2f}", styles["Normal"]))
table_data = [data.columns.tolist()] + data.head(10).values.tolist()

table = Table(table_data)
table.setStyle(TableStyle([
    ('BACKGROUND', (0,0), (-1,0), colors.grey),
    ('TEXTCOLOR', (0,0), (-1,0), colors.whitesmoke),
    ('GRID', (0,0), (-1,-1), 1, colors.black),
    ('ALIGN', (0,0), (-1,-1), 'CENTER'),
]))
elements.append(table)

pdf.build(elements)

print("Startup_Profit_Report.pdf generated successfully!")
```
